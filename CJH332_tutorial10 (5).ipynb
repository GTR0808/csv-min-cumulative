```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# assignment = pd.read_csv("./Data/Assignment.csv")
assignment = pd.read_csv("Assignment.csv")
assignment.columns = assignment.columns.str.lower()
print("Columns in dataset:", assignment.columns.tolist())
```

```python
# Part 2: Visualize the relationship
fig, ax = plt.subplots()
ax.scatter(assignment["genea_expression"], assignment["cell_viability"])

ax.set_xlabel("geneA_expression")
ax.set_ylabel("cell_viability")
ax.set_title("geneA_expression vs. cell_viability")
plt.show()
```

```python
# Part 3: Fit a linear regression model
lr = smf.ols("cell_viability ~ genea_expression", data=assignment).fit()
```

```python
# Your summary for part 3 here
print(lr.summary())
```
